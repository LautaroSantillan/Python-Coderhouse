#Listas

```python
#Listas en Python: heterogeneas (list)
nombres = ["Lautaro", "Nazareno", "Tomas"]
lista = [100, 8.8, "Hello World!", 'c', -10]
type(nombres)
type(lista)
```

```python
#Indexacion
print(nombres[0])
print(lista[2])
print(lista[-1])
```

```python
#Concatenación => Junta los elementos
lista_Grande = nombres + lista
print(lista_Grande)
```

```python
#Mutabilidad
lista[0] = '¿?'
print(lista)
```

```python
#Slicing
print(lista_Grande[1:4])
```

```python
#Clear
lista_Grande.clear() # => Elimina los elementos de una lista
print(lista_Grande)
```

```python
#Append => Añadir un valor al final
lista_Grande.append("Hi everyone!")
print(lista_Grande)
nombres.append('Marcelo')
print(nombres)
```

```python
#Longitud de lista
print(len(nombres))
print(len(lista))
print(len(lista_Grande))
```

```python
#Eliminar un elemento
nombres.pop() # => elimina el ultimo elemento (default) y te dice cual es
#nombres.pop(0) => eliminaria 'Lautaro' por indice
```

```python
print(nombres)
```

```python
#Contar un elemento => count
numbers = [1, 2, 3, 1, 4, 5, 6, 1, 7, 8, 9]
numbers.count(1)
```

```python
#Encontrar el indice => index
numbers.index(6) # => busca el item y nos dice en qué indice se encuentra
```

#Tuplas

```python
#Crear tuplas => El tamaño y la cantidad de elementos ya queda asi, es estatico
numeros = (20, 8, -2, 12)
tupla = ("Hola", "como", "estas", "?")
tupla_vacia = (2,)
```

```python
mi_var = 'Una variable'
datos = (1, -5, 123, 34, "Una cadena", "Otra cadena", mi_var)
type(datos)
```

```python
#Indexación
print(datos[-1])
print(datos[4])
print(datos[0])
```

```python
#Concatenación =>  NO FUNCIONA APPEND
numeros = (5, 6, 7, 8, 9, 10)
(1, 2, 3, 4) + numeros
```

```python
#Mutabilidad => No va a funcionar porque las tuplas son inmutables
mi_tupla = (1,2,3,4)
mi_tupla[2] = 5
```

```python
#Slicing
info = ('Lautaro', 19, 'Ezequiel', -88.6, 'Santillan', 1.83)
info[::2]
#info[1::2] => (19, -88.6, 1.83)
```

```python
#Borrar valores => También es la forma de instanciar una tupla vacía
vocales = ('a', 'e', 'i', 'o', 'u')
vocales = ()
print(vocales)
```

```python
#Longitud de una tupla
print(len(tupla))
print(len(datos))
print(len(info))
```

```python
#Count
numbers = [1, 2, 3, 1, 4, 5, 6, 1, 7, 8, 9, 1, 10, 11, 12]
numbers.count(1)
```

```python
#Index
numbers.index(9)
```
