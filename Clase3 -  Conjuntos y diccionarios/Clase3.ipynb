#CONJUNTOS Y DICCIONARIOS

#Conjuntos

Colección no ordenadas de objetos unicos

```python
#Conjuntos en Python --- set (no tiene orden y no permite elementos repetidos)
planetas = {'Tierra', 'Mercurio', 'Urano', 'Jupiter', 'Venus', 'Marte', 'Saturno', 'Neptuno', 'Tierra', 'Urano'}
#print(planetas[-1]) #No existe el orden --- no hay indexacion
print(planetas)
```

```python
#Conversión de listas a sets o de listas a tuplas --- set()
lista1 = (1, 2, 3, 4, 5)
conjunto_1 = set(lista1)
print(conjunto_1)
lista1 = list(conjunto_1)
print(lista1)
```

```python
#Heterogéneos
conjunto2 = {30, 'treinta', 30.0}
```

* Diferencias: listas y conjuntos ---> el orden y la repeticion de elementos
* Tuplas y conjuntos ---> el orden, la repetición y además la mutabilidad

```python
#Agregar un nuevo item: add()
conjunto2.add(30.25) # => Lo añade en cualquier lado porque no hay orden
print(conjunto2)
conjunto2.add(-30)
print(conjunto2)
```

```python
#Añadir múltiples elementos: update()
conjunto2.update([3, "Chau"]) # => puede ser una tabla o tupla
print(conjunto2)
```

```python
#Longitud del conjunto: len()
print(len(planetas))
```

```python
#Eliminar todos los valores: clear()
conjunto2.clear()
print(conjunto2)
```

```python
#Una forma de eliminar item: discard() --- indicar el item a eliminar
conjunto3 = {1, 2, 3, 5}
conjunto3.discard('python') # => si le indico un item que no existe, no le hace caso
conjunto3.discard(5)
print(conjunto3)
```

```python
#Otra forma de eliminar item: remove() --- indicar el item a eliminar
conjunto3.add(5)
print(conjunto3)
# conjunto3.remove(10) # => si le indico un item que no existe, me tira un error
conjunto3.remove(5)
print(conjunto3)
```

```python
#Borrar un elemento y mostrarlo: pop()
conjunto4 = {'L', 'M', 'M', 'J', 'V', 'S', 'D', 'Fin'}
print(conjunto4.pop())
print(conjunto4)
```

#Diccionarios

Colección no ordenada de objetos ordenados por *clave:valor* = (key:value)

Las claves **suelen ser int o string**, aunque **cualquier otro objeto inmutable** puede actuar como una clave. Los valores. por el contrario, pueden ser de **cualquier tipo**, inclusio otros diccionarios

```python
#Creación de un diccionario
paises = {'Argentina':'Buenos Aires', 'Paraguay':'Asuncion', 'Uruguay':'Montevideo', 'Brasil':'Brasilia', 'Chile':'Santiago'}
```

```python
#Obteniendo el valor = value
print(paises['Argentina'])
```

```python
datos = {"Nombre":"Lautaro", "Apellido":"Santillan", "Edad": 19, "Peso": 87.5, "Altura": 1.83}
#print(datos)
print("Nombre del alumno:", datos['Apellido'], datos['Nombre'])
```

```python
#Mutabilidad
print(paises['Chile'])
paises['Chile'] = 'Santiago de Chile'
print(paises['Chile'])
```

```python
#Asignación
print(datos['Edad'])
datos['Edad'] += 1
print("La edad de alumno es de:" ,datos['Edad'])
```

##Funciones de diccionarios

```python
#Add => No existe una funcion tal pero se puede realizar asi:
numeros = {'uno': 1, 'dos': 2, 'tres': 3, 'cuatro': 4}
print(numeros)
numeros['cinco'] = 5
print(numeros)
```

```python
#Update
numeros.update({'seis': 6, 'siete': 7, 'ocho': 8, 'nueve': 9})
print(numeros)
num_dic = dict(diez=10)
numeros.update(num_dic)
print(numeros)
```

```python
#Longitud del diccionario
len(paises)
```

```python
#Del => Elimina el item del dict, sin modificar el resto del dict
#del numeros['once'] # => salta un error
del numeros['nueve']
del numeros['tres']
print(numeros)
```

```python
#In => Sirve para ver si pertenece un elemento a un dict
'tres' in numeros # => arroja false
9 not in numeros # => arroja true
'ocho' in numeros
```

```python
#Clear => Borrar todos los valores de un dict
print(numeros)
numeros.clear() # => Tambien se puede hacer numeros = {}
print(numeros)
```
