##Consigna Sets

Crear un conjunto en Python que posea los siguientes elementos:
* Países: Inglaterra, USA, México.
* Posteriormente agrega nuestro set de países, los elementos de: Islandia, Italia, Argentina, USA y Portugal.
* Elimina a los países: Chile e Italia.

Pregunta: ¿Qué pasa si queremos eliminar al país Chile utilizando el método remove?, ¿Qué pasó con el element de USA?


```python
paises = {'Inglaterra', 'USA', 'México'}
paises.update(['Islandia', 'Italia', 'Argentina', 'Usa', 'Portugal'])
print(paises)
paises.discard('Chile')
paises.discard('Italia')
print(paises)
# paises.remove('Chile') => lo que sucede con remove es que si no existe el item, salta un error de ejecucion
# => Lo que paso con USA es que no se agrega porque en un set no se repiten elemento
```

##Consigna Dicts

Escribir un programa que le solicite al usuario su nombre, edad, dirección y que, posteriormente, lo muestre por pantalla:

Ejemplo del output solicitado:
* Juan tiene 25 años, y vive en Carrera 7 - Bogotá


```python
nombre = input("Ingrese su nombre: ")
edad = int(input("Ingrese su edad: "))
direccion = input("Ingrese su dirección y ciudad: ")
usuario = {'nombre': nombre, 'edad': edad, 'dirección': direccion}
print(usuario['nombre'], 'tiene', usuario['edad'], 'años, y vive en', usuario['dirección'])
```
