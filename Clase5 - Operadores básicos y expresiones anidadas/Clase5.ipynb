##Operadores Logicos: AND OR NOT

- **and**: El resultado final es verdaderos SOLO SI todo es True (si hay 1 o mas False se vuelve todo False)
- **or**: El resultado final es verdadero SI AL MENOS 1 condición es verdadero (solo será False si todos son False)
- **not**: El resultado final será falso si la condición no es verdadera

```python
1 > 0 and 5 < 10
```

```python
(1 == 0) and (0 < 100) and (12 == 12)
```

```python
1 == 0 or 0 < 100 or 12 == 12
```

```python
not True
```

```python
not ("casa" == "CASA")
```

##Operadores de asignación

```python
numero = 8
```

```python
#Suma en asignación
numero += 1
print(numero)
```

```python
#Resta en asignación
numero -= 3
print(numero)
```

```python
#Producto en asignación
numero *= 5
print(numero)
```

```python
#División en asignación
numero /= 3
print(numero)
```

```python
#Potencia en asignación
numero **= 3
print(numero)
```

```python
#Modulo en asignación
numero %= 2
print(numero)
```
