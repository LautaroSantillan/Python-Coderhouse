Crear una variable que almacene si se cumplen todas las condiciones

A partir de dos variables llamadas **NOMBRE** y **EDAD**, debes crear una variable que almacene si se cumplen todas las siguientes condiciones, encadenando operadores lógicos en una sola línea:

* **NOMBRE** sea diferente de cuatro asteriscos “****”
* **EDAD** sea mayor que 5 y a su vez menor que 20
* Que la **longitud** de NOMBRE sea mayor o igual a 4  pero a la vez menor que 8
* **EDAD** multiplicada por 3 sea mayor que 35


Desde un input conseguir las variables:

nombre = INPUT!!!

edad = INPUT!!!!


```python
#Nombre
print('Las condiciones deben ser: ser distinto a "****" y que la longitud del nombre sea mayor o igual a 4 y menor a 8')
resultado = True
while resultado == True:
  nombre = input('Ingrese un nombre: ')
  long_Edad = len(nombre)
  if(nombre != "****") and (long_Edad >= 4 and long_Edad < 8):
    print('Cumple con las condiciones')
    break
  else:
    print('El nombre no cumple con todos los requisitos, vuelva a ingresar')
```

```python
#Edad
print('Las condiciones son ser mayor a 5 y menor a 20, y que multiplicado con 3 sea mayor a 35')
resultado = True
while resultado == True:
  edad = int(input('Ingrese una edad: '))
  if (edad > 5 and edad < 20) and (edad * 3 > 35):
    print('Cumple con las condiciones')
    break
  else:
    print('La edad no cumple con todos los requisitos, vuelva a ingresar')
```
