###Argumentos indeterminados

```python
def sumar(*args): # una cantidad indeterminada de argumentos
  return sum(args)
```

```python
sumar(5, 1, 98, 7, 54, 2)
```

```python
def suma(**kwargs):
  s = 0
  for key, value in kwargs.items():
    print(key, "-", value)
    s += value
  return s
```

```python
suma(a=5, b=12, c=-6)
```

###Funciones recursivas

```python
# función recursiva sin retorno
def cuenta_regresiva(numero):
  numero -= 1
  if numero > 0:
    print(numero)
    cuenta_regresiva(numero)
  else:
    print("Boooom!!")
  print("Fin de la función n°", numero)
```

```python
cuenta_regresiva(5)
```

```python
# función recursiva con retorno
def factorial(numero):
  print("Valor inicial -> ", numero)
  if numero > 1:
    numero = numero * factorial(numero-1)
  print("Valor final -> ", numero)
  return numero
```

```python
factorial(5)
```

###Funciones integradas

```python
# int
x = "25"
print(type(x))
x = int("25")
print(type(x))
```

```python
# float
x = float(2.5)
print(x)
print(type(x))
```

```python
# En estos casos no funcionaría
x = float(25L)
print(x)
print(type(x))
x = float(“2”)
print(x)
print(type(x))
x = float(“2.5”)
print(x)
print(type(x))
```

```python
# str
z = str(12)
print(z)
print(type(z))
# => se puede pasar un float a un str
z = str(3.2)
print(z)
print(type(z))
# => se puede pasar un complex a un str
z = str(2.3+0j)
print(z)
print(type(z))
# => se puede pasar un booleano a un str
z = str(True)
print(z)
print(type(z))
```

```python
# round => La función round() redondea un número flotante a una precisión dada en dígitos decimal
# La precisión tal vez sea negativa.
x = round(2.3)
print(x)
y = round(2.6)
print(y)
```
