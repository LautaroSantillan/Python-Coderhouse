#Conversiones tipos de datos

Pasaremos de milímetros a metros según el parámetro de la función

###Descripción de la actividad.

Realiza una función que, dependiendo de los parámetros que reciba, convierta a milímetros o a metros.  
1. Si recibe un argumento, supone que son milímetros y convierte a metros, centímetros y milímetros.
2. Si recibe 3 argumentos, supone que son metros, centímetros y milímetros y los convierte a milímetros.


```python
def conversion(*args):
  if len(args) == 1:
    print(f"Ingreso 1 argumento => Milimetros: {args[0]}")
    mm = args[0] # valor total de los milimetros
    metros = mm//1000 # total en metros
    mm -= metros * 1000 # actualizar milimetros
    cm = mm//10 # total en cm
    mm -= cm * 10 # actualizar milimetros
    print(f"Metros: {metros} --- Centimetros: {cm} --- Milimetros: {mm}")
  elif len(args) == 3:
    print("Ingreso 3 argumentos")
  else:
    print("Error: Ingrese 1 argumento o 3 argumentos")
```

```python
conversion(72975)
```
