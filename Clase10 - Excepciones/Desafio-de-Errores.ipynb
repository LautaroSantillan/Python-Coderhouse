Retorna None cuando sucede el fallo

###Descripción de la actividad

En la función de nuestro ejercicio hay un fallo potencial, averigua cuándo sucede y retorna el valor None en ese caso especial, en cualquier otro caso devuelve el resultado normal de la función.

`def dividir(a, b):`

`return a/b`

**Pista:** Valor indeterminado


```python
#Puede haber 2 errores
# ValueError => Si ingresan algo que no sea un int
# ZeroDivisionError => Si se ingresa un cero en una variable
def dividir():
  try:
    x = int(input("Ingrese un numero: "))
    y = int(input("Ingrese otro numero: "))
    return print(f"La división entre ambos numeros es: {x/y}")
  except ValueError:
    print("Por favor, ingresa un número entero!")
  except ZeroDivisionError:
    print("No se puede dividir con un cero(0)")
  except:
    print("Ocurrió un error desconocido!")
```
