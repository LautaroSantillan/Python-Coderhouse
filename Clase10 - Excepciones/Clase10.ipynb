```python
#try-except
try:
  x = int(input("Ingrese un numero: "))
  y = int(input("Ingrese otro numero: "))
  print(f"La división entre ambos numeros es: {x/y}")
except ValueError:
  print("Por favor, ingresa un número entero!")
except ZeroDivisionError:
  print("No se puede dividir con un cero(0)")
except:
  print("Ocurrió un error desconocido!")
```

###Verificación de Datos - Else - Finally

```python
while True:
  try:
    a = float(input("Ingrese un numero: "))
    b = float(input("Ingrese otro numero: "))
    print(a + b)
  except: # se ejecuta solo si alguna linea del try falla
    print("Ha ocurrido un error. Tienes que introducir 2 numeros")
  else: # se ejecuta si es que todas las lineas en el try son correctas
    print("Datos correctos")
    break
```

```python
#Validación de datos - else
while True:
  try:
    edad = int(input("Ingrese su edad: "))
  except ValueError:
      print("Por favor, ingrese un numero entero como edad!")
  else:
      print("Gracias, su edad ha sido almacenada")
      break
```

```python
#Validación de datos - finally
while True:
  try:
    edad = int(input("Ingrese su edad: "))
  except ValueError:
    print("Por favor, ingrese un numero entero como edad!")
  else:
    print("Gracias, su edad ha sido almacenada")
    break
  finally:
    print("Fin del try-except")
```

```python
try:
  #Ponemos todas las lineas que nos pueden generar un error potencial
  pass
except:
  #Una acción o un mensaje en caso de que se genere un error en cualquier linea del try
  pass
else: #Lo contrario al except
  #Una acción o mensaje SOLO SI ha habido un error potencial (todas las lineas se ejecutaron OK)
  pass
finally:
  #Siempre se ejecutara --- haya habido un error o no durante el try
  pass
```
