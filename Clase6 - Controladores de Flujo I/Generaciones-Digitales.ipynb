###Descripción de la actividad.

Escribir un programa que indique la generación correspondiente para un año de nacimiento indicado.

**Importante:** Para los años que no pertenezcan a ninguna generación, se deberá colocar: “No existe generación asociada”

- Generación Silenciosa (1920-1938)
- Niños de la Posguerra (1939-1945)
- Baby Boomer (1946-1964)
- Generación X (1965-1979)
- Generación Y (1980-2000)
- Generación Z (2001-2010)
- Generación Alfa (2011-2024)


```python
year_of_birth = int(input('Ingrese su año de nacimiento: '))
if 1920 <= year_of_birth <= 1938:
  print("Usted pertenece a la Generación Silenciosa")
elif 1939 <= year_of_birth <= 1945:
  print("Usted pertenece a la generación Niños de la Posguerra")
elif 1946 <= year_of_birth <= 1964:
  print("Usted pertenece a la generación Baby Boomer")
elif 1965 <= year_of_birth <= 1979:
  print("Usted pertenece a la Generación X")
elif 1980 <= year_of_birth <= 2000:
  print("Usted pertenece a la Generación Y")
elif 2001 <= year_of_birth <= 2010:
  print("Usted pertenece a la Generación Z")
elif 2011 <= year_of_birth <= 2024:
  print("Usted pertenece a la Generación Alfa")
else:
  print("No existe generación asociada (1920 al 2024)")
```
