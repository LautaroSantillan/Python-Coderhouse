#Controladores de Flujo I

```python
#format String
nombre = input("Ingrese tu nombre: ")
edad = int(input("Ingrese tu edad: "))

print(f"Hola, me llamo {nombre} y tengo {edad} años")
```

```python
#if
nombre = input("Ingrese tu nombre: ")
edad = int(input("Ingrese tu edad: "))
if edad >= 18:
  print(f"{nombre}: Mayor de edad")
```

```python
#else
nombre = input("Ingrese tu nombre: ")
edad = int(input("Ingrese tu edad: "))
if edad >= 18:
  print(f"{nombre}: Apto para conducir")
else: #por el contrario (else)
  print(f"{nombre}: No apto para conducir")
```

```python
#if anidados
x = 18
if x > 10:
  print("Por encima de 10")
  if x > 20:
    print("Y también por encima de 20")
  else:
    print("Pero no por encima de 20")
else:
  print("Pero no por encima de 10")
```

```python
#elif
comando = input("Ingrese un comando en mayúsculas: ").upper()
if comando == "ENTRAR":
  print('Bienvenido al sistema')
elif comando == "SALUDO":
  print('Hola!! Como estas?')
elif comando == "SALIR":
  print('Saliendo del sistema')
else:
  print('No se reconoce el comando')
```
