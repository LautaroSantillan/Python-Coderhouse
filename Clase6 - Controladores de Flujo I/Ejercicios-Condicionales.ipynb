3. Escriba un programa que pida ingresar un número entero y diga la
cantidad de cifras que tiene. En el caso que el número tenga 4
cifras o más mostrar un mensaje diciendo: “Número muy grande.”

```python
num = int(input("Ingrese un numero menor a 4 cifras: "))
if num >= 1000:
  print("Numero muy grande")
elif 100 <= num < 1000:
  print("El número tiene 3 cifras")
elif 10 <= num < 100:
  print("El número tiene 2 cifras")
else:
  print("El número tiene 1 cifra")
```

4. Escriba un programa que indique si un número es múltiplo de 3 o
no

```python
num = int(input("Ingrese un numero: "))
if num % 3 == 0:
  print(f"El numero ingresado ({num}) es multiplo de 3")
else:
  print(f"El numero ingresado ({num}) no es multiplo de 3")
```

5. Preguntar la edad a una persona y decidir en qué grado va a estar.
* 5 o menos --- Aún no estás en el colegio
* 6-11 --- 1ro - 6to primaria
* 12-16--- 1ro - 5to secundaria
* 17 o más --- Estás en la universidad


```python
edad = int(input("Ingrese su edad: "))
if 0 < edad <= 5:
  print("Aún no estás em el colegio")
elif 6 <= edad <= 11:
  print(f"Estás en grado {edad-5} de primaria")
elif 12 <= edad <= 16:
  print(f"Estás en grado {edad-11} de secundaria")
elif edad >= 17 and edad < 100:
  print("Estás en la universidad")
else:
  print("Ingrese una edad valida")
```

6. Pedir 3 números y ordenarlos de menor a mayor.

```python
num1 = int(input("Ingrese un numero a elección: "))
num2 = int(input("Ingrese otro numero a elección: "))
num3 = int(input("Ingrese otro numero a elección: "))
if (num1 < num2) and (num1 < num3):
  if(num2 < num3):
    print(num1, num2, num3)
  else:
    print(num1, num3, num2)
elif (num2 < num1) and (num2 < num3):
  if(num1 < num3):
    print(num2, num1, num3)
  else:
    print(num2, num3, num1)
else:
  if(num1 < num2):
    print(num3, num1, num2)
  else:
    print(num3, num2, num1)
```

7. Pedir ingresar los lados de un triángulo e indicar de qué tipo es:
* Isósceles: 2 lados iguales
* Equilátero: 3 lados iguales
* Escaleno: 0 lados iguales


```python
lado1 = int(input("Ingrese el primer lado del triangulo: "))
lado2 = int(input("Ingrese el segundo lado del triangulo: "))
lado3 = int(input("Ingrese el tercer lado del triangulo: "))
if (lado1 == lado2) and (lado2 == lado3):
  print("Triángulo Equilátero")
elif (lado1 == lado2) or (lado1 == lado3):
  print("Triángulo Isósceles")
elif (lado2 == lado1) or (lado2 == lado3):
  print("Triángulo Isósceles")
else:
  print("Triángulo Escaleno")
```

8. Un postulante a un empleo, realiza un test de capacitación, se
obtuvo la siguiente información: cantidad total de preguntas que
se le realizaron y la cantidad de preguntas que contestó
correctamente. Se pide confeccionar un programa que ingrese los
dos datos por teclado e informe el nivel del mismo según el
porcentaje de respuestas correctas que ha obtenido, y sabiendo
que:
- Nivel máximo: Porcentaje>=90%.
- Nivel medio: Porcentaje>=75% y <90%.
- Nivel regular: Porcentaje>=50% y <75%.
- Fuera de nivel: Porcentaje<50%.

```python
cant_preguntas = int(input("Ingrese la cantidad total de preguntas que se le realizaron al postulante: "))
cant_respuestas = int(input("Ingrese la cantidad de preguntas que contestó correctamente el postulante: "))
porcentaje = (cant_respuestas * 100) / cant_preguntas
if 90 <= porcentaje <= 100:
  print("El postulante alcanzó el nivel máximo")
elif 75 <= porcentaje < 90:
  print("El postulante alcanzó el nivel medio")
elif 50 <= porcentaje < 75:
  print("El postulante alcanzó el nivel regular")
else:
  print("El postulante está fuera de nivel")
```
