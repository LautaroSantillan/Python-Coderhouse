###Herencia Multiple


```python
class Animal:
  def __init__(self, especie, edad):
    self.especie = especie
    self.edad = edad

  def hablar(self):
    print("El animal ha hablado!!")

  def moverse(self):
    print("El animal se ha movido!!")

  def describeme(self):
    print("Soy un animal de tipo:", type(self).__name__)
```

```python
class Cuadrupedo:
  def hablar(self):
    print("El Cuadrupedo ha hablado!!")

  def correr(self):
    print("El Cuadrupedo corre usando sus 4 patas")
```

```python
class Perro(Animal, Cuadrupedo):
  def hablar(self):
    print("Guau!")
```

```python
Pluto = Perro("Bloodhound", 7)
Pluto.hablar()
```

###MRO

```python
class A:
  def imprimir(self):
    print("Hola, soy A")

class B:
  pass

class C:
  def imprimir(self):
    print("Hola, soy C")

class D:
  def imprimir(self):
    print("Hola, soy D")
```

```python
class Hija(B, C, A):
  pass
```

```python
a = Hija()
a.imprimir()
```

```python
print(Hija.__mro__) #Method Order Resolution
```
