###Clases y Objetos

```python
class Perro:

  especie = "Mamifero" #Atributo de clase (es constante para todos los objetos)

  def __init__(self, nombre, raza, size):
    self.name = nombre #a veces el complemento del self no es igual a lo que dice a su derecha
    self.breed = raza
    self.size = size
    self.edad = 0
  #Métodos
  def ladrar(self):
    return "Guau"
```

```python
snoopy = Perro(nombre = "Snoopy", raza = "Beagle", size = "Mediano")
mi_perro = Perro("Lola", "Coker", "Mediano")
```

```python
#Atributos de instancia (pueden variar dependiendo del objeto)
print(f"El perro que he creado se llama {snoopy.name}")
print(f"Mi perra se llama {mi_perro.name}")
```

```python
#Atributo de clase (constantes para todos los objetos creados)
Perro.especie #nombre_objeto.nombra_atributo
```

```python
snoopy.ladrar() #nombre_objeto.nombra_metodo(Argumentos Necesarios)
```

```python
class Tarjeta:
  empresa = "VISA"
  forma = "rectangular"

  def __init__(self, numero, fecha, banco, nombre):
    self.numero = numero
    self.fecha = fecha
    self.banco = banco
    self.titular = nombre

  def activar(self):
    print(f"Hola {self.titular}!! Tu tarjeta del banco {self.banco} ha sido activada")

  def pagar(self, monto, lugar):
    print(f"Has realizado un consumo por el monto de {monto} en {lugar}")
```

```python
tarjeta_1 = Tarjeta(4001, "01/29", "BBVA", "Pepe Python")
tarjeta_2 = Tarjeta(4002, "05/27", "Santander Rio", "Alumno Pepito")
```

```python
#Accediendo a un método
tarjeta_1.activar()
```

```python
tarjeta_2.activar()
```

```python
#Accediendo a un atributo
tarjeta_2.numero
```

```python
tarjeta_1.pagar(7000, "Burguer King")
```

###Composición de clases

```python
class Motor:
  def __init__(self, tipo, potencia):
    self.tipo = tipo
    self.potencia = potencia
```

```python
class Coche:
  def __init__(self, marca, modelo, year):
    self.marca = marca
    self.modelo = modelo
    self.año = year
    self.motor = Motor("electrico", 180) #Este atributo es un objeto de otro tipo de clase
    #Tiene sus propios atributos
```

```python
match5 = Coche("Meteoro", "Match 5", 1980)
```

```python
print(match5.motor.tipo) #nombre_objeto.nombre_atributo_del_objeto_original.nombre_atributo_objeto_2
```

###Anidación de clases

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  class Employee:
    def __init__(self, position, salary):
      self.position = position
      self.salary = salary
```

```python
person1 = Person("Lautaro", 20)
```

```python
#La persona creada antes, recien ahora también es empleado
emp1 = person1.Employee("Sereno", 1000)
```

```python
print(person1.name)
print(emp1.salary)
```

###Métodos especiales

```python
class Ejemplo:
  def __init__(self, atr1, atr2):
    self.atr1 = atr1
    self.atr2 = atr2

  def __str__(self):
    return "He creado un objeto de tipo Ejemplo"

  def __len__(self):
    return 10

  def __getitem__(self, pos):
    return self.atr1[pos]
```

```python
ejemplo1 = Ejemplo("Hola", 123)
```

```python
print(ejemplo1)
```

```python
len(ejemplo1)
```

```python
ejemplo2 = Ejemplo("Chau", "Bye!")
```

```python
len(ejemplo2)
```

```python
ejemplo3 = Ejemplo([1, 2, 3, 4, 5], "Hi")
ejemplo3[2]
```

```python
ejemplo3[4]
```

###Encapsulamiento

```python
class Alumno:
  def __init__(self, nombre, nota):
    self.nombre = nombre
    self.__nota = nota #Hacemos que no se pueda saber sobre este atributo desde
    # afuera de la clase (no se puede leer el atributo porque es privado)
    self.edad = 0

  def imprimir_nota(self):
    print(f"El estudiante {self.nombre} tiene una nota de {self.__nota}")

  def get_nota(self): #getter
    return self.__nota

  def set_nota(self, valor): #setter
    self.__nota = valor
```

```python
alumno_1 = Alumno(nombre = "Lautaro", nota = 8)
```

```python
print(alumno_1.nombre)
print(alumno_1.nota)
```

```python
alumno_1.__resultado()
```

```python
alumno_1.__nota = 6
```

```python
alumno_1.imprimir_nota()
#Como el atributo nota es un atributo privado, no se puede alterarlo desde afuera
#por eso que cuando pedimos la nota sigue estando la anterior
```

```python
#Para setear un atributo privado, es asi:
alumno_1.set_nota(6)
alumno_1.imprimir_nota()
```
