```python
#Programación imperativa
lista = [1, 2, 3, 4, 5]
for x in range(0, 5): #x tomaría indices del 0 al 4
  lista[x] *= 2
print(lista)
```

```python
#Programación declarativa --- funcional
def duplicar(x):
  return x*2

lista = (1,2,3,4,5)

resultado = map(duplicar, lista) #map(funcion, lista)

print(tuple(resultado))
```
