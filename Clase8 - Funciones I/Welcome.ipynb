#Welcome

Escribir una función a la que se le pase una cadena del nombre de una ciudad <ciudad> y muestre por pantalla el saludo ¡hola bienvenidx a <nombre>!.


```python
def welcome(city):
  return print(f"¡Hola bienvenido/a a {city}!")
```

```python
city = input("Ingrese la ciudad en la que se encuentra ahora: ")
welcome(city)
```
