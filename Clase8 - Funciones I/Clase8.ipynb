```python
def mul_7(): # se define la función
  numero = int(input("Ingrese un numero: "))
  print(f"El numero multiplicado por 7 es {numero*7}")
```

```python
mul_7() # invocando la función
```

```python
def capitalize_ult():
  palabra = input("Ingrese la palabra: ")
  palabra_final = palabra[:-1] # corta la palabra antes de la última letra
  ultima_letra = palabra[-1] # obtiene la última letra
  palabra_final += ultima_letra.upper() # le agrego la última letra en MAYUS a la palabra final
  print(palabra_final)
```

```python
capitalize_ult()
```

```python
# función con parametro y return
def saludar_con_nombre(nombre):
  saludando = f"Hola {nombre}, como estas?"
  return saludando
```

```python
saludar_con_nombre("Lautaro") # invocando a la función con un argumento
```

```python
# funcion con doble parametro
def sum(num1, num2):
  return num1 + num2
```

```python
x = sum(8, 5)
print(x)
```
