###Descripción de la actividad.

Utilizando todo lo que sabes sobre cadenas, listas y sus métodos internos, transforma este texto:

**gordon lanzó su curva&strawberry ha fallado por un pie! -gritó Joe Castiglione&dos pies -le corrigió Troop&strawberry menea la cabeza como disgustado… -agrega el comentarista**

**Transforma el texto en:**

Gordon lanzó su curva...
- Strawberry ha fallado por un pie! -gritó Joe Castiglione.
- Dos pies -le corrigió Troop.
- Strawberry menea la cabeza como disgustado… -agrega el comentarista.

_Lo único prohibido es modificar directamente el texto_


```python
texto = "gordon lanzó su curva&strawberry ha fallado por un pie! -gritó Joe Castiglione&dos pies -le corrigió Troop&strawberry menea la cabeza como disgustado… -agrega el comentarista"
texto_sin_yComercial = texto.replace('&', ' ', 3)
print(texto_sin_yComercial)

#Primera oracion
print(texto_sin_yComercial.find('st'))
parte_1 = texto_sin_yComercial[:22:1]
p1 = parte_1.capitalize() + "..."

#Segunda oracion
print(texto_sin_yComercial.find('dos'))
parte_2 = texto_sin_yComercial[22:79]
print(parte_2.find('Joe'))
print(parte_2.find('Cas'))
nombre = parte_2[41:44]
print(nombre)
apellido = parte_2[45:-1]
print(apellido)
cap = parte_2.capitalize()
print(cap)
lista2 = cap.split()
lista2[7] = nombre
lista2[-1] = apellido
print(lista2)
p2 = '- ' + ' '.join(lista2)

#Tercera oracion
print(texto_sin_yComercial.rfind('str'))
parte_3 = texto_sin_yComercial[79:107]
print(parte_3.find('Tr'))
nombre = parte_3[22:-1]
print(nombre)
cap = parte_3.capitalize()
print(cap)
lista3 = cap.split()
lista3[-1] = nombre
print(lista3)
p3 = '- ' + ' '.join(lista3)

#Cuarta oracion
parte_4 = texto_sin_yComercial[107::]
p4 = '- ' + parte_4.capitalize()
```

```python
print(p1)
print(p2)
print(p3)
print(p4)
```
