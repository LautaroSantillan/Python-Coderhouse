Analizar e interpretar los siguientes conjuntos en Python

Ejecuta el código para analizar cada una de las salidas!


```python
#Ejemplo 1:
my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])
my_new_set = my_set_1.union(my_set_2)
print(my_new_set)
# => Lo que paso aca es que se realizo una unión entre el set 1 ^ el set 2
# y el 3 no se repitió porque estamos en conjuntos
```

```python
#Ejemplo 2:
my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])
my_new_set = my_set_1.intersection(my_set_2)
print(my_new_set)
# => Aca lo que sucede es que arroja '{3}' porque es el elemento en común
# que hay entre ambos sets
```

```python
#Ejemplo 3:
my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])
my_new_set = my_set_1.difference(my_set_2)
print(my_new_set)
# => Arroja '{1, 2}' porque son elementos no comunes entre ambos sets
```
