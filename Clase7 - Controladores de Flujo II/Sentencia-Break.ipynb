#Sentencia break

###Descripción de la actividad.

Observa el código y busca comprender qué sucedió en este caso:

    x = 5

    while True:
    
        x -= 1
        
        print(x)
        
        if x == 0:
            
            break
        
        print("Fin del bucle")


- Lo que sucede es que el while es verdadero y entra, resta 1 a **x** y lo muestra. Y la condición de fin del bucle es cuando **x = 0**.
- Pero el "error" está en que cuando se resta 1 a x, muestra también el **print("Fin del bucle")**, cuando no sucedió aún el fin de ciclo, se da por un error de identación.
