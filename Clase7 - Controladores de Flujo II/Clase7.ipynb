##While

```python
#Repetir una acción de forma eficaz
numero = 1
while numero <= 10: #Mientras numero sea menor o igual a 10, entonces
  print(numero)
  numero += 1
```

```python
#Operadores de asignación
a = 1
a += 10 # a = a + 10
a -= 7 # a = a - 7
print(a)
```

```python
#Solución al Bucle Infinito
n = 5
while (n > 0):
  print(n)
  n -= 1
```

##While - Break - Else

```python
chance = 1
while chance <= 3:
  text = input("Escribe la contraseña: ")
  if text == "4658720":
    print("Ok, lo lograste en el intento", chance)
    break
  chance += 1
else:
  print("Se acabaron tus intentos. Celular bloqueado!")
```

```python
x = 5
while True: # esta condición NO se va a detener nunca
  x -= 1
  print(x)
  if (x == 0):
    break # freno para salir del bucle infinito
print("Fin del bucle")
```

##For

```python
#Numeros del 1 al 10 --- for each (por cada)
for numero in range(1, 11):
  print(numero)
```

```python
lista_nombres = ["Antonio", "Gabriel", "Mateo"]
for nombre in lista_nombres:
  print(nombre)
```

```python
for x in range(1, 11): #range(inicio, final+1)
  print(x)
```

```python
#Desafio --- Imprimir numeros del 1 al 100 (solo numeros pares)
for x in range(1, 101):
  if x % 2 == 0:
    print(x)
```

```python
for multiplo in range(1, 51):
  if multiplo % 5 == 0:
    print(multiplo)
```

##While (Continue y Pass)

```python
#continue
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
#pass
n = 0
while n < 10:
  n += 1
  if n == 2:
    pass #placeholder
  print(f"n vale {n}")
```

#For (Continue y Pass)

```python
#continue
for i in range(1, 11):
  if i == 3:
    continue
  else:
    print(i)
```

```python
#pass
for numero in range(10):
    if numero == 2:
        pass
    elif numero == 8:
        break
    else:
        print("Se terminó de iterar y numero vale: ", numero)
```
