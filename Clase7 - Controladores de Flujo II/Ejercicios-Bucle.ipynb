###3. Pedir ingresar un número y mostrar el factorial de dicho número.

Objetivo: El factorial de un número es el número multiplicado por
todos los números positivos menores a él.

Ejemplo: 5! = 5x4x3x2x1 = 120


```python
num = int(input("Ingrese un numero: "))
fact = 1
while num > 0:
  fact *= num
  num -= 1
print(f"El factorial del numero es: {fact}")
```

###4. Pedir un número entero de 4 dígitos y escribir el número al revés
(tratar de hacerlo sin usar cadenas de texto)

```python
num = int(input("Ingrese un numero de 4 digitos: "))
if 1000 <= num <= 9999:
  numero_invertido = 0
  for x in range(4):
    digito = num % 10
    numero_invertido = numero_invertido * 10 + digito
    num //= 10
else:
    print("Ingrese un número de 4 dígitos")
print(f"El numero invertido es: {numero_invertido}")
```

###5. Pedir un número e imprimir la cantidad de divisores que tiene.


```python
num = int(input("Ingrese un numero: "))
i = num
cont = 0
while i > 0:
  if num % i == 0:
    cont += 1
    print(f"Divisor {cont}: {num/i}")
  i -= 1
print(f"El numero tiene {cont} divisores")
```

###6. Pedir un número y analizar si dicho número es primo.

Objetivos: Un número es primo si solo tiene dos divisores: 1 y el mismo
número.

Ejemplo: 7 es primo pero 8 no porque tiene como divisores a 1,2,4 y 8.


```python
num = int(input("Ingrese un numero: "))
cont = 0
for i in range(1, num+1):
  if num % i == 0:
    cont += 1
if cont == 2:
  print(f"El numero {num} es primo")
else:
  print(f"El numero {num} no es primo")
```

###7. Pedir ingresar una palabra e imprimir la cantidad de vocales que tiene.

```python
word = input("Ingrese una palabra: ")
x = list(word)
cant_letras = len(word)
cont = 0
for i in range(cant_letras):
  if x[i].upper() == 'A' or x[i].upper() == 'E' or x[i].upper() == 'I' or x[i].upper() == 'O' or x[i].upper() == 'U':
    cont += 1
print(f"La cantidad de vocales que tiene la palabra {word} es de: {cont}")
```

###8. Vamos a dibujar un árbol de navidad. Construir un programa que pida ingresar la altura de un árbol y devuelva el árbol impreso con asteriscos
(pueden usar cualquier otro símbolo).

```python
alt = int(input("Ingrese la altura que quiera para su arbol: "))
if alt > 0:
  raiz = alt
  for x in range(alt+1):
    print(' '*(alt) + '*'*((x*2)-1))
    alt -= 1
  print(' '*(raiz-1) + '*')
else:
  print("No quiere un arbolito")
```
